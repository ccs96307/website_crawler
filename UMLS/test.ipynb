```python
with open("source.html", "r", encoding="utf-8") as f:
    data = f.read()

len(data)
```

```python
soup = BeautifulSoup(data, "html.parser")
lines = soup.prettify()
print(len(lines))
```

```python
pattern = re.compile(r"<a id=\"http.+?href=\"(.+?)\".+?</a>")
name_urls = pattern.findall(data)
name_urls = [url for url in name_urls if "callback" not in url]
len(name_urls)
```

```python
import html
import urllib.parse

prefix = "https://bioportal.bioontology.org/"
postfix = "#mappings"

urls = []
for url in name_urls:
    url = prefix + html.unescape(urllib.parse.unquote(url)) + postfix
    urls.append(url)
    print(url)
    
    # if "RID34785" in url:
    #     print(url)
```
