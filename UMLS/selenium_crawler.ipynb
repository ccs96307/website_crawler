```python
from selenium import webdriver
from bs4 import BeautifulSoup

url = "https://bioportal.bioontology.org/ontologies/RADLEX/?p=classes&conceptid=http%3A%2F%2Fradlex.org%2FRID%2FRID15849"

# Init
driver = webdriver.Chrome(executable_path="PATH_TO_CHROME_DRIVER")
driver.get(url)

# Expand all <li> element on the page
driver.execute_script("""
function expandListItem(listItemElements, index) {
  if (index >= listItemElements.length) {
    return;
  }

  var listItem = listItemElements[index];
  var trigger = listItem.querySelector(".trigger");
  if (trigger) {
    trigger.click();

    // Detect all <li> elements
    var childListItems = listItem.querySelectorAll("li.folder-close");

    // Wait for loading
    setTimeout(function() {
      // Expand recursively
      expandListItem(childListItems, 0);

      // Process <li> element in the same level
      setTimeout(function() {
        expandListItem(listItemElements, index + 1);
      }, 2000);
    }, 2000);
  } else {
    // If there is not trigger, continue to process the next element
    expandListItem(listItemElements, index + 1);
  }
}

// Get the top <li> element
var topLevelListItemElements = document.querySelectorAll("li.folder-close");

if (topLevelListItemElements.length > 0) {
  expandListItem(topLevelListItemElements, 0);
}
""")

# Get the page source code
html = driver.page_source

# Parsing
soup = BeautifulSoup(html, "html.parser")

# Find and process <li> element
list_items = soup.find_all("li")

for item in list_items:
    pass

# Close
driver.quit()
```
